### **DATA READING**

```python
df = spark.read.format("parquet").load(f"abfss://source@databricksetehimanshi.dfs.core.windows.net/{p_file_name}")
df.display()
```

```python
df= spark.readStream.format("cloudFiles")\
         .option("cloudFiles.format","parquet")\
         .option("cloudFiles.schemaLocation",f"abfss://bronze@databricksetehimanshi.dfs.core.windows.net/checkpoint_{p_file_name}")\
         .load(f"abfss://source@databricksetehimanshi.dfs.core.windows.net/{p_file_name}")
```

### **DATA WRITING**

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation",f"abfss://bronze@databricksetehimanshi.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .option("path",f"abfss://bronze@databricksetehimanshi.dfs.core.windows.net/{p_file_name}")\
    .trigger(once=True)\
    .start()
```
